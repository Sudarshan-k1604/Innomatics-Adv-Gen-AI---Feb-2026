```python
# Problem Statement:
# Given a sentence, identify all unique words and display:
# 1. The count of unique words
# 2. The unique words themselves

user_line = input("Enter sentence: ")

split_data = user_line.split()

distinct_terms = set(split_data)

result_count = len(distinct_terms)

print("Unique words count", result_count)
print("Unique words", distinct_terms)
```

```python
# Problem Statement:
# Store employee names and their salaries, and determine which employee has the highest salary.

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest = max(employees, key=employees.get)

salary = employees[highest]

print("Highest Salary ", highest, "-", salary)
```

```python
# Problem Statement:
# Find the maximum and minimum values from a list of numbers.

numbers = [45, 22, 89, 10, 66]

big = numbers[0]
small = numbers[0]

for val in numbers:
    if val > big:
        big = val
    if val < small:
        small = val

print("List:", numbers)
print("Max:", big)
print("Min:", small)
```

```python
# Problem Statement:
# Count how many products have a price greater than 1000.

prices = [450, 1200, 899, 1500, 300]

count = 0

for val in prices:
    if val > 1000:
        count += 1

print("Products above 1000:", count)
```

```python
# Problem Statement:
# Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).

attendance = ["P", "P", "A", "P", "P"]

present = 0
total = len(attendance)

for mark in attendance:
    if mark == "P":
        present += 1

percent = (present / total) * 100

print("Attendance Percentage:", percent)
```

```python
# Problem Statement:
# Count how many times each character appears in a given string.

text = "pythonp"

count = {}

for ch in text:
    if ch in count:
        count[ch] += 1
    else:
        count[ch] = 1

print(count)
```

```python
# Problem Statement:
# Check whether a specific key exists in a dictionary.

staff = {
    "Employee": "Suresh",
    "Age": 28,
    "City": "Mysore"
}

key = "Employee"

if key in staff:
    print("Employee exists")
else:
    print("Employee does not exist")
```

```python
# Problem Statement:
# Calculate the average marks of a student from a list of marks.

marks = [70, 75, 80, 65]

total = 0

for val in marks:
    total += val

avg = total / len(marks)

print("Average Marks:", avg)
```

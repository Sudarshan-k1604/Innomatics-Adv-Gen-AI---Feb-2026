```python
"""
Problem 1: Post Engagement Analyzer
"""
def analyze_post(like_data):
    total_count = 0
    for value in like_data:
        total_count += value

    if total_count >= 1000:
        result = "Viral Post"
    else:
        result = "Normal Engagement"

    print("Total Likes:", total_count)
    print("Post Status:", result)


analyze_post([200, 300, 250, 300])
```

```python
"""
Problem 2: Medicine Stock Alert
"""
def check_medicine(stock_qty):
    if stock_qty < 10:
        state = "Low Stock Alert"
    else:
        state = "Stock Sufficient"

    print("Medicine Stock:", stock_qty)
    print("Status:", state)


check_medicine(6)
```

```python
"""
Problem 3: Rainfall Adequacy Checker
"""
def rainfall_check(rain_list, required_level):
    total_rain = 0
    for rain in rain_list:
        total_rain += rain

    avg_rain = total_rain / len(rain_list)

    if avg_rain >= required_level:
        condition = "Adequate Rainfall"
    else:
        condition = "Inadequate Rainfall"

    print("Average Rainfall:", avg_rain)
    print("Rainfall Status:", condition)


rainfall_check([60, 80, 75, 73], 70)
```

```python
"""
Problem 4: Duplicate Account Detection
"""
def detect_duplicates(user_list):
    unique_set = set(user_list)

    if len(unique_set) != len(user_list):
        flag = "Yes"
    else:
        flag = "No"

    print("Duplicate Accounts Found:", flag)


detect_duplicates(["user1", "user2", "user3", "user1"])
```

```python
"""
Problem 7: Application Health Checker
"""
def system_health(error_total):
    if error_total == 0:
        health_state = "Healthy"
    elif error_total <= 5:
        health_state = "Minor Issues"
    else:
        health_state = "Critical Issues"

    print("Error Count:", error_total)
    print("System Status:", health_state)


system_health(7)
```

```python
"""
Problem 8: Transaction Limit Checker
"""
def transaction_check(amount_value):
    daily_cap = 50000

    if amount_value <= daily_cap:
        decision = "Approved"
    else:
        decision = "Rejected"

    print("Transaction Amount:", amount_value)
    print("Transaction Status:", decision)


transaction_check(60000)
```

```python
"""
Problem 9: Attendance Eligibility System
"""
def attendance_check(attendance_marks):
    present_days = 0

    for mark in attendance_marks:
        present_days += mark

    percentage = (present_days / len(attendance_marks)) * 100

    if percentage >= 75:
        exam_status = "Eligible"
    else:
        exam_status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", exam_status)


attendance_check([1,1,1,1,0,1,1,1,0,1])
```

```python
"""
Problem 10: Smart Electricity Bill Analyzer
"""
def electricity_bill(unit_count):
    total_charge = 0

    if unit_count <= 100:
        total_charge = unit_count * 3
    elif unit_count <= 200:
        total_charge = (100 * 3) + ((unit_count - 100) * 5)
    else:
        total_charge = (100 * 3) + (100 * 5) + ((unit_count - 200) * 7)

    if total_charge < 500:
        usage_type = "Low Usage"
    elif 500 <= total_charge <= 1500:
        usage_type = "Moderate Usage"
    else:
        usage_type = "High Usage"

    return total_charge, usage_type


bill_amount, usage_label = electricity_bill(250)

print("Total Bill:", bill_amount)
print("Usage Category:", usage_label)
```

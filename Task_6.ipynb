```python
"""Smart Parking Lot Management System"""
def parking_manager(max_slots, movement_logs):
    current_count = 0
    highest_count = 0

    for action in movement_logs:
        if action == "IN":
            current_count += 1
        elif action == "OUT" and current_count > 0:
            current_count -= 1

        if current_count > highest_count:
            highest_count = current_count

    if current_count > max_slots:
        status_msg = "Capacity Exceeded"
    else:
        status_msg = "Available"

    print("Currently Parked Vehicles:", current_count)
    print("Parking Status:", status_msg)


parking_manager(50, ["IN","IN","IN","OUT","IN","IN","OUT"])




```

```python
"""Online Food Delivery Time Estimator"""
def delivery_eta(distance_km, traffic_type, weather_type):
    base_time = distance_km * 5 # 5 minutes per km

    if traffic_type == "High":
        base_time += 10
    elif traffic_type == "Medium":
        base_time += 5

    if weather_type == "Rainy":
        base_time += 5
    elif weather_type == "Storm":
        base_time += 10

    print("Estimated Delivery Time:", base_time, "minutes")


delivery_eta(8, "High", "Rainy")
```

```python
"""Movie Theatre Seat Occupancy Analyzer"""
def theatre_analysis(total_capacity, booking_list):
    booked_total = 0

    for seat in booking_list:
        booked_total += seat

    occupancy_rate = (booked_total / total_capacity) * 100

    if occupancy_rate == 100:
        show_state = "Housefull"
    elif occupancy_rate >= 75:
        show_state = "Almost Full"
    else:
        show_state = "Seats Available"

    print("Occupancy:", str(int(occupancy_rate)) + "%")
    print("Show Status:", show_state)


theatre_analysis(200, [1]*150)
```

```python
"""Cloud Server Load Classification"""
def server_load_checker(cpu_values):
    total_cpu = 0

    for reading in cpu_values:
        total_cpu += reading

    avg_cpu = total_cpu / len(cpu_values)

    if avg_cpu < 50:
        server_state = "Normal"
    elif avg_cpu <= 80:
        server_state = "Warning"
    else:
        server_state = "Critical"

    print("Average CPU Load:", str(int(avg_cpu)) + "%")
    print("Server Status:", server_state)


server_load_checker([45, 60, 70, 85, 90])
```

```python
"""Smart Classroom Resource Usage Monitor"""
def classroom_monitor(resource_data):
    overused_list = []
    alert_flag = "No"

    for resource, hours in resource_data.items():
        if hours > 8:
            overused_list.append(resource)
            alert_flag = "Yes"

    print("Overused Resources:", ", ".join(overused_list))
    print("Energy Alert:", alert_flag)


classroom_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

```python
"""Online Event Registration Capacity Controller"""
def registration_control(event_limit, total_users):
    if total_users <= event_limit:
        confirmed = total_users
        waiting = 0
        reg_status = "Open"
    else:
        confirmed = event_limit
        waiting = total_users - event_limit
        reg_status = "Closed"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waiting)
    print("Registration Status:", reg_status)


registration_control(100, 105)
```
